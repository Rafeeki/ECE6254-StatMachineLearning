```python
# Standard Library imports
%matplotlib inline
import math, numpy
import matplotlib.pyplot as plt
```

```python
# Problem 2d
B = [3, 4, 5, 6]
for b in B:
    m = 10**b
    X = numpy.zeros(m)
    for i in range(0,m):
        x = 1 * numpy.random.randn(m) + 0 # Normal(0,1)
        if x.max() > X.max(): X = x
    plt.hist(X)
    plt.title("Histogram of X with m = " + str(m))
    plt.show()
```
